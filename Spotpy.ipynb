# Clusters

```python
from sklearn.cluster import KMeans
import numpy as np
```

```python
numeric_audio_features = total_audio_features_df.select_dtypes(include = np.number)
```

```python
numeric_audio_features
```

```python
inertia_list = list()
clusters = list(range(2,10))

for c in clusters:
    kmeans = KMeans(n_clusters=c, random_state=0).fit(numeric_audio_features)
    inertia_list.append(round(kmeans.inertia_, 2))
```

```python
clusters
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
fig = plt.figure(figsize = (12, 8))
plt.plot(clusters, (inertia_list)) 
plt.title('Elbow plot', fontsize = 15)
plt.show()
```

```python
final_kmeans = KMeans(n_clusters=4, random_state=0).fit(numeric_audio_features)
```

```python
numeric_audio_features['cluster'] = final_kmeans.predict(numeric_audio_features)
```
